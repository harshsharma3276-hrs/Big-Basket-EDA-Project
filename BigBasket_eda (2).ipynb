### Top & least sold products (using rating as proxy)

```python
print("Top Products:")
print(df.sort_values(by='rating', ascending=False).head())

print("\nLeast Products:")
print(df.sort_values(by='rating').head())
```

### Discount Calculation

```python
df['discount'] = df['market_price'] - df['sale_price']
df[['product','sale_price','market_price','discount']].head()
```

### Missing Values

```python
df.isnull().sum()
```

```python
# Handling null values

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns


df[num_cols] = df[num_cols].fillna(df[num_cols].mean())


for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

### Visualization

```python
import matplotlib.pyplot as plt

plt.scatter(df['sale_price'], df['rating'])
plt.xlabel('Sale Price')
plt.ylabel('Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df['category'].value_counts().plot(kind='bar')
plt.title("Products per Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['brand'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Brands")
plt.xlabel("Brand")
plt.ylabel("Number of Products")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['sale_price'], bins=40)
plt.title("Sale Price Distribution")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot([df['sale_price'], df['market_price']], labels=['Sale Price', 'Market Price'])
plt.title("Price Comparison")
plt.ylabel("Value")
plt.show()
```

```python
import seaborn as sns

numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Only)")
plt.show()
```
